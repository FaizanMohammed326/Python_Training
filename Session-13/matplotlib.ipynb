```python
"""
Matplotlib is an amazing visualization library in Python for 2D plots of arrays. Matplotlib is a multi-platform data
visualization library built on NumPy arrays

 Matplotlib consists of several plots like line, bar, scatter, histogram etc.


"""
from matplotlib import pyplot as plt
x = [1, 5, 8, 9, 2]
y = [5, 9, 0, 3, 7]
plt.plot(x,y)
plt.show()
```

```python
"""
This function accepts parameters that enables us to set axes scales and format the graphs.
To plot only the markers, you can use shortcut string notation parameter 'o', which means 'rings'.
"""
plt.plot(x,y,'o')
plt.show
```

```python
"""

If we do not specify the points in the x-axis, they will get the default values 0, 1, 2, 3,"""
ypoints = [3, 8, 1, 10]

plt.plot(ypoints)
plt.show()
```

```python
"""
Markers:
    You can use the keyword argument marker to emphasize each point with a specified marker:
"""
plt.plot(ypoints, marker = 'o')
plt.show()
```

```python
plt.plot(ypoints, marker = '*')
plt.show()
```

```python
"""Format Strings fmt
You can use also use the shortcut string notation parameter to specify the marker.

This parameter is also called fmt, and is written with this syntax:

marker|line|color"""
plt.plot(ypoints, 'o:r')
plt.show()
```

```python
"""
'-'    Solid line
':'    Dotted line
'--'   Dashed line
'-.'   Dashed/dotted line
"""
plt.plot(ypoints, '+-.b')
plt.show()
```

```python
"""
You can use the keyword argument markersize or the shorter version, ms to set the size of the markers:
You can use the keyword argument markeredgecolor or the shorter mec to set the color of the edge of the markers:
You can use the keyword argument markerfacecolor or the shorter mfc to set the color inside the edge of the markers:
"""
plt.plot(ypoints, marker = 'o', ms = 20, mec = 'r', mfc = 'green')
plt.show()
```

```python
"""
You can use the keyword argument linestyle, or shorter ls, to change the style of the plotted line:
"""
plt.plot(ypoints, ls = 'dashed')
plt.show()
```

```python
"""
You can use the keyword argument color or the shorter c to set the color of the line:
"""
plt.plot(ypoints, color = 'r')
plt.show()

"""
You can use the keyword argument linewidth or the shorter lw to change the width of the line.
"""
plt.plot(ypoints, lw = '20.5')
plt.show()
```

```python
"""
You can plot as many lines as you like by simply adding more plt.plot() functions:

Example
"""
y1 = [2,8,5,7,1]
y2 = [9,12,3,5,1]
plt.plot(y1)
plt.plot(y2)
plt.show()
```

```python
x1 = [0,2,4,6,8]
x2 = [8,3,7,2,9]
plt.plot(x1, y1, x2, y2)
plt.show()
```

```python
"""
Create Labels for a Plot
With Pyplot, you can use the xlabel() and ylabel() functions to set a label for the x- and y-axis.
With Pyplot, you can use the title() function to set a title for the plot.
"""
plt.plot(x1, y1)
plt.title('Delivery')
plt.xlabel('Price')
plt.ylabel('Distance')
plt.show()
```

```python
"""
With Pyplot, you can use the grid() function to add grid lines to the plot.
You can use the axis parameter in the grid() function to specify which grid lines to display.

"""
plt.plot(x1, y1)
plt.title('Delivery')
plt.xlabel('Price')
plt.ylabel('Distance')
plt.grid(axis = 'x')
plt.show()
```

```python
"""
You can also set the line properties of the grid, like this: grid(color = 'color', linestyle = 'linestyle', linewidth = number).
"""
plt.plot(x1, y1)
plt.title('Delivery')
plt.xlabel('Price')
plt.ylabel('Distance')
plt.grid(color = 'red', linestyle = '-.', linewidth = 1.5)
plt.show()
```

```python
"""
Display Multiple Plots
With the subplots() function you can draw multiple plots in one figure:
The subplots() function takes three arguments that describes the layout of the figure.
first and second argumens are the number of rows and columns iof graphs
third column represents the index of the current plot
"""
x = [2,10,40,80]
y = [1,13,34,9]
plt.subplot(1,2,1)
plt.plot(x,y)
x = [10,20,30,40]
y = [25,15,12,34]
plt.subplot(1,2,2)
plt.plot(x,y)
plt.show()
```

```python
"""
You can add a title to the entire figure with the suptitle() function:
"""
x = [2,10,40,80]
y = [1,13,34,9]
plt.subplot(1,2,1)
plt.plot(x,y)
plt.title('goals')


x = [10,20,30,40]
y = [25,15,12,34]
plt.subplot(1,2,2)
plt.plot(x,y)
plt.title('assists')

plt.suptitle('Stats')
plt.show()
```

```python
"""Ticks are the values used to show specific points on the coordinate axis. It can be a number or a string.
axis            Tells which axis to operate
reset           If True, set all parameters to default
direction       Puts the ticks inside or outside or both
length          Sets ticks length
width           Sets ticks width
rotation        Rotates ticks wrt the axis
colors          Changes tick color
pad             Distance in points between tick and label

Default plot Example:
"""
x = [5, 10, 15, 20, 25, 30, 35, 40, 45, 50]
y = [1, 4, 3, 2, 7, 6, 9, 8, 10, 5]
 
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
 
plt.show()
```
